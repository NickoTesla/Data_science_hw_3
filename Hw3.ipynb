```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import normalize
```

```python
# 1. Функція гіпотези
def hypothesis(w, X):
    return np.dot(X, w)
```

```python
# 2. Функція втрат
def compute_loss(w, X, y):
    m = len(y)
    h = hypothesis(w, X)
    loss = (1/(2*m)) * np.sum((h - y)**2)
    return loss
```

```python
# 3. Один крок градієнтного спуску
def gradient_descent_step(w, X, y, alpha):
    m = len(y)
    h = hypothesis(w, X)
    gradient = (1/m) * np.dot(X.T, (h - y))
    w = w - alpha * gradient
    return w
```

```python
# 4. Градієнтний спуск
def gradient_descent(X, y, alpha, num_iterations):
    m, n = X.shape
    w = np.zeros(n)
    for _ in range(num_iterations):
        w = gradient_descent_step(w, X, y, alpha)
    return w
```

```python
# 5. Аналітичне рішення
def analytical_solution(X, y):
    return np.linalg.inv(X.T @ X) @ X.T @ y
```
